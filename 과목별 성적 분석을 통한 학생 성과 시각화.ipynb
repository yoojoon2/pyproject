# 과목별 성적 분석을 통한 학생 성과 시각화

```python
import pandas as pd  # 데이터프레임 다룰 때 사용
import numpy as np  # 숫자 계산과 랜덤 값 생성용
import matplotlib.pyplot as plt  # 그래프 그리기용

# 한글 폰트 설정 
plt.rcParams['font.family'] = 'Malgun Gothic'

class StudentScoreAnalysis:
    def __init__(self):
        # '학생1'부터 '학생20'까지 이름 만들기
        self.students = [f'학생{i}' for i in range(1, 21)]

        # 과목 리스트
        self.subjects = ['수학', '영어', '과학']

        # 50~100 사이 난수로 생성 (20명 x 3과목)
        self.scores = np.random.randint(50, 101, size=(20, 3))

        # 데이터프레임 생성 (학생 이름: index, 과목명: column)
        self.df = pd.DataFrame(self.scores, index=self.students, columns=self.subjects)

    def subject_average(self):
        # 과목별 평균 계산 (열 기준 평균)
        subject_avg = self.df.mean()

        # 과목별 평균 점수 막대그래프
        plt.figure(figsize=(8, 5))
        subject_avg.plot(kind='bar', color=['red', 'green', 'blue'])
        plt.title('과목별 평균 점수')
        plt.xlabel('과목')
        plt.ylabel('평균 점수')
        plt.ylim(0, 100)
        plt.grid(axis='y')
        plt.tight_layout()
        plt.show()

    def top5_students(self):
        # 학생별 평균 점수 계산 (행 기준 평균)
        student_avg = self.df.mean(axis=1)

        # 평균 점수 높은 순으로 정렬한 후 상위 5명 추출
        top5 = student_avg.sort_values(ascending=False).head(5)

        # 상위 5명 막대그래프
        plt.figure(figsize=(8, 5))
        top5.plot(kind='bar', color='gray')

        plt.title('평균 성적 상위 5명')
        plt.xlabel('학생 이름')
        plt.ylabel('평균 점수')
        plt.ylim(0, 100)
        plt.grid(axis='y')
        plt.tight_layout()
        plt.show()

# 실행
if __name__ == '__main__':
    Analysis = StudentScoreAnalysis()
    Analysis.subject_average()
    Analysis.top5_students()
```
